```python
"""5. Leer un número entero y determinar el número de cincos que están consecutivos
en un número. Por ejemplo, en 5512355551 hay 6 cincos consecutivos. Ahora, en 
234555234512 solo hay 3 cincos consecutivos."""

#definición y salida en pantalla mensaje de bienvenida al programa.
welc = 'BIENVENIDO AL PROGRAMA QUE CALCULA LA CANTIDAD DE 5 CONSECUTIVOS.'

print(welc.center(100))

#Lectura de datos desde el teclado. Se pide al usuario ingresar un número
#entero. En caso de no ingresar un número, se pide ingresar de nuevo el dato.
while True:
    try:
        num = int(input("Digite un número entero: "))  
        break                                           
    except ValueError:                                  
        print("Lo que ingresó no es un número entero")  

#Condicional si número ingresado es negativo. En ese caso, se multiplica por 
#(-1) cambiando así su signo.
if num < 0:
    num = (-1)*num

#Se definen las variables. 
cont5 = 0         #Contador de cincos.
                                       
ant = 0           #Número anterior.
                                       
cant5 = 0         #Contador de cincos seguidos.                                      

#Se define un ciclo while, a través del cual, se itera a través del número,
#descomponiéndolo en dígitos mediante una división entera entre 10 y % para 
#hallar el residuo. la variable dig va tomando el valor del dígito que resulta 
#de este proceso.
while num > 0:
    dig = num % 10
    
    #Se define un condicional if. Mediante este, valiéndose de las propiedades 
    #de la división entera, se comprueba que los dígitos sean cincos. Si lo 
    #son, el contador de cincos aumenta.
    if dig == 5:
        cont5 += 1
        #Se define un condicional if anidado. Mediante este, 
        #se comprueba que el dígito hallado y el anterior sean cincos. Si lo 
        #son, al contador de cincos consecutivos se le suma el contador de 
        #cincos.
        if dig == 5 and ant == 5:
            cant5 += cont5
            cont5 = 0
            
    else:
        ant = dig
        cont5 = 0
    ant = dig
    num = num // 10
#Salida en pantalla de la cantidad de cincos.
print('En el número hay', cant5, 'cincos consecutivos.')    
```

```python
"""6. Leer un número entero y determinar si la suma de sus dígitos es un número de
Fibonacci."""

#definición y salida en pantalla mensaje de bienvenida al programa.
welc = 'Hola, con este programa podrá saber si la suma de los dígitos de un \
número es un número de la serie fibonacci.'

print(welc.center(100))

#Lectura de datos desde el teclado. Se pide al usuario ingresar un número
#entero. En caso de no ingresar un número, se pide ingresar de nuevo el dato.
while True:
    try:
        num = int(input("Digite un número entero: "))
        break                                           
    except ValueError:                                  
        print("Lo que ingresó no es un número entero")
        
#Condicional si número ingresado es negativo. En ese caso, se multiplica por 
#(-1) cambiando así su signo.
if num < 0: 
    num = (-1)*num
    
#Se define la variable suma como la que al final será la suma de dígitos.
suma = 0

#Se definde un ciclo while que permite hallar los dígitos del número ingresado
#y sumarlos
while num > 0:
    resi = num % 10
    suma += resi
    num = num // 10

print("La suma de los digitos es: ", suma)

#Se define la función fibonacci como una sucesión de números, resultado de la 
#suma de los dos anteriores miembros de la sucesión, partiendo desde 0 y 1.    
def fibonacci(suma):                                      
    a = 0                                                        
    b = 1
    fib = a+b
    while fib < suma:
        fib = a+b
        a = b
        b = fib
        if fib == suma:
            return True
    return False

#Definición de las salidas en pantalla, si se cumple que suma hace parte de la
#sucesión o no.
if fibonacci(suma) or suma == (1) or suma == 0:
    print('Este número es un número de la serie Fibonacci.')
else:
    print('Este número es NO pertenece a la serie Fibonacci.')                 
         
#listo, pero buscar forma diferente
```

```python
"""7. Leer un número entero positivo y determine si existen en el número dígitos 
repetidos. Reporte cuales son los dígitos repetidos."""

#definición y salida en pantalla mensaje de bienvenida al programa.
welc = 'BIENVENIDO AL PROGRAMA QUE BUSCA DÍGITOS REPETIDOS'

print(welc.center(100))

#Lectura de datos desde el teclado. Se pide al usuario ingresar un número
#entero positivo. En caso de no ingresar un número o que este sea negativo, 
#se pide ingresar de nuevo el dato.
while True:
    try:
        num = int(input("Digite un número entero positivo: "))
        if num > 0:
            break
        else: 
            print('El número debe ser mayor que 0. Inténtelo de nuevo.')
    except ValueError:
        print('Debe ingresar un entero positivo. Inténtelo de nuevo.')

#Se definen       
tupdig={}           #Una tupla vacía que podrá o no contener unos.
                
listarep=[]         #Una lista que contendrá los dígitos repetidos.

listadig=[]         #Una lista que contendrá los dígitos del número.

#Se define un ciclo while, a través del cual, se itera a través del número,
#descomponiéndolo en dígitos mediante una división entera entre 10 y hallando 
#el residuo mediante %. Los dígitos se van añadiendo a la lista "listadig".
while num != 0:
   
        k=num%10
        num=num//10 
        listadig.append(k)

#Se define un ciclo for para iterar a través de "listadig" y hallar los 
#elementos repetidos
for x in listadig:
    if x not in tupdig:
         tupdig[x] = 1
    else:
        #Se añaden los dígitos repetidos a "listarep".
        if tupdig[x] == 1:
            listarep.append(x)
            
#Salida en pantalla dependiendo de si hay o no dígitos repetidos.
if listarep >= [0]:
    print('El número contiene dígitos repetidos. La lista de estos es: \n' 
f'{listarep}')
else:
    print('El número no contiene dígitos repetidos.')
```

```python
"""9. Hacer un programa que lea las coordenadas (x1,y1,r1) y (x2,y2,r2) que 
corresponden al centro y al radio de dos círculos. Lea un punto de coordenadas 
(a,b). Determina si (a,b) está contenido: a) dentro del círculo 1; b) dentro 
del cí­rculo 2; c) dentro de ambos cí­rculos; d) fuera de ambos cí­rculos. 
Recordemos que el cí­rculo es el lugar geométrico de los puntos del plano cuya 
distancia a otro punto fijo, llamado centro, es menor o igual que una cantidad 
constante, llamada radio."""

#definición y salida en pantalla mensaje de bienvenida al programa.
welc = 'Hola, este programa le permitirá saber si un punto se encuentra\
o no dentro de dos cí­rculos ingresados por usted.'

print(welc.center(100))

#Se importa la librería math.
import math as mt

#Lectura de datos desde el teclado. Se pide al usuario ingresar centros y 
#radios de 2 círculos y un punto a su gusto. En caso de no ingresar números o 
#radios negativos se pide ingresar de nuevo el dato.
while True:
    
    try:
        x1 = float(input('Ingrese la coordenada en x del centro del' 
                         ' primer círculo: '))
        y1 = float(input('Ingrese la coordenada en y del centro del' 
                         ' primer círculo: '))
        
        x2 = float(input('Ingrese la coordenada en x del centro del' 
                         ' segundo círculo: '))
        y2 = float(input('Ingrese la coordenada en y del centro del' 
                         ' segundo círculo: '))
        
        a = float(input('Ingrese la coordenada en x de un punto: '))
        b = float(input('Ingrese la coordenada en y de un punto: '))
        break
    except ValueError:
        print ('Los valores ingresados deben ser números reales. Inténtelo de' 
               ' nuevo.')
while True:
    try:
        r1 = float(input('Ingrese el radio del primer círculo: '))
        if r1 > 0: break
        else:
            print('El radio del círculo debe ser un número positivo')
    except ValueError:
        print('El radio del círculo debe ser un número')
              
while True:
    try:
        r2 = float(input('Ingrese el radio del segundo círculo: '))
        if r2 > 0: break
        else:
            print('El radio del círculo debe ser un número positivo')
    except ValueError:
        print('El radio del círculo debe ser un número')
        
#ditancia punto al centro del círculo 1.
dp1=mt.sqrt(((a-x1)**2)+((b-y1)**2))

#ditancia punto al centro del círculo 2.
dp2=mt.sqrt(((a-x2)**2)+((b-y2)**2))

#Salida en pantalla de los resultados.
if dp1 <= r1:
    print('El punto de coordenadas (a,b) está contenido en el círculo 1.')
    
if dp2 <= r1:
    print('El punto de coordenadas (a,b) está contenido en el círculo 2.')
    
if dp1 <= r1 and dp2 <= r1:
    print('Así pues, el punto está contenido por los 2 círculos')
    
if dp1 > r1 and dp2 > r1:
    print('Ninguno de los círculos definidos contienen al punto de coordenadas' 
          '(a,b).')
```

```python
"""10. Sin utilizar el método .upper() para cadenas, escribir un programa que lea 
una cadena de texto y la imprima en mayúscula. Tenga en cuenta que el programa 
debe aceptar las letras con tilde, la ñ y la ü."""

#definición y salida en pantalla mensaje de bienvenida al programa.
welc = 'BIENVENIDO AL PROGRAMA QUE PERMITE PONER UN TEXTO EN MAYÚSCULA'

print(welc.center(100))

#Lectura de datos desde el teclado. Se pide al usuario ingresar una cadena.
txt = (input("Ingrese una cadena de texto: "))
          
#Se define cadena de letras minúsculas.   
let_min='a b c d e f g h i j k l m n ñ o p q r s t u v w x y z á é í ó ú ü'

#Se define cadena de letras mayúsculas.   
let_cap='A B C D E F G H I J K L M N Ñ O P Q R S T U V W X Y Z Á É Í Ó Ú Ü'

#Se crean listas con las cadenas anteriores.
l_or=let_min.split(' ')
l_rep=let_cap.split(' ')

#Se define un ciclo for para iterar a través del texto ingresado.
#Se reemplazan los caracteres hallados en el texto y que estén en la lista de
#minúsculas por su equivalente en la lista de mayúsculas.
for i in range(len(l_or)):
    if str(l_or[i]) in txt:
        txt = txt.replace(str(l_or[i]) , str(l_rep[i]))
    else: 
        txt = txt
        
#Se imprime el texto en mayúsculas.
print('El texto ingresado, ahora en mayúscula es: ', txt)
```

```python
"""11. Escribir un programa que reciba una cadena de texto y reporte:
a) cuántas letras vocales en mayúscula se entraron;
b) cuántas letras con tilde se entraron (minúsculas y mayúsculas);
c) cuántos dígitos se entraron,
d) cuántos espacios se entraron;
e) cuántas palabras reservadas se entraron.
Utilizar para ello los métodos de cadenas especializados como .isupper(), 
.isdigit(), etc. y la palabra reservada in."""

#definición y salida en pantalla mensaje de bienvenida al programa.
welc = """Este programa le permitirá ingresar una cadena y saber: \n
a) cuántas letras vocales en mayúscula se entraron.
b) cuántas letras con tilde se entraron (minúsculas y mayúsculas).
c) cuántos dígitos se entraron.
d) cuántos espacios se entraron.
e) cuántas palabras reservadas se entraron.
"""

print(welc)

#Se importa la librería keyword
import keyword

#Lectura de datos desde el teclado. Se pide al usuario ingresar una cadena.
txt = (input("Ingrese una cadena de texto: "))

#Se definen:
vow = 'AEIOU'           #Cadena que contiene las vocales mayúsculas

til_cap = 'ÁÉÍÓÚ'       #Cadena que contiene Las vocales mayúsculas tildadas

til_min = 'áéíóú'       #Cadena que contiene Las vocales minúsculas tildadas

esp = ' '               #Cadena que contiene un esoacio

key = (keyword.kwlist)  #De la librería que se importó con anterioridad, se 
                        #define key como la lista de palabras restringidas

#Se hace al texto ingresado una cadena
lista_cadena = txt.split(' ')

#Se definen los contadores:
cont_vow = 0            #Se define el contador de vocales mayúsculas

cont_til = 0            #Se define el contador de letras tildadas

cont_dig = 0            #Se define el contador de dígitos

cont_esp = 0            #Se define el contador de espacios

cont_res = 0            #Se define el contador de palabras restringidas

#Se itera a través de la cadena, modificando los contadores dependiendo del 
#caso
for i in range(len(txt)):
    if txt[i] in vow or txt[i] in til_cap:
        cont_vow += 1 
        
    if txt[i] in til_cap or txt[i] in til_min:
        cont_til += 1
        
    if txt[i].isdigit() is True:
        cont_dig += 1
        
    if txt[i] in esp:
        cont_esp += 1 

for j in lista_cadena:
    if j in key:
        cont_res+=1

#Salida en pantalla del resultado
print('la cadena ingresada contiene: ' '\n' f'{cont_vow} vocales en mayúscula'
      '\n' f'{cont_til} letras con tilde entre mayúsculas y minúsculas' '\n'
      f'{cont_dig} dígitos' '\n' f'{cont_esp} espacios' '\n' f'{cont_res} '
      'palabras reservadas')
```

```python
"""12. Hacer un programa que lea dos palabras (podrían estar en mayúscula o 
minúscula) y determine cual está primero en el diccionario. El programa debe 
soportar letras con tilde, la ü y la ñ."""

#definición y salida en pantalla mensaje de bienvenida al programa.
welc = 'BIENVENIDO AL PROGRAMA ORGANIZADOR DE PALABRAS.'

print(welc.center(100))

#Se define una cadena de caracteres aceptados
cad_car = ('abcdefghijklmnñopqrstuvwxyzABCDEFGHIJKLMNÑÑOPQRSTUVWXYZáéíóúü\
ÁÉÍÓÚÜ')
car_perm = set(cad_car)

print('Recuerde que las palabras que ingrese pueden contener los caracteres:\n'
      ,list(cad_car), end='')

#Lectura de datos desde el teclado. Se pide al usuario ingresar dos palabras.
#Si estas palabras contienen caracteres que no están en la cadena de caracteres 
#permitidos se pide que sea ingresada de nuevo.
while True:
    try:                                                                 
        word_1 = input('\nIngrese la primera palabra: ')
      
        if set(word_1).issubset(car_perm):
            break
        else:
            print('Sólo puede ingresar letras en la cadena. Inténtelo de \
nuevo')
    except ValueError:
        print ('Ingrese de nuevo la cadena')
        
word_1 = word_1.title()
        
while True:
    try:                                                                 
        word_2 = input('Ingrese la segunda palabra: ')
        
        if set(word_2).issubset(car_perm):
            break
        else:
            print('Sólo puede ingresar letras en la cadena. \
Inténtelo de nuevo')
            
    except ValueError:
        print ('Ingrese de nuevo la cadena')

#Ambas palabras son transformadas a título.
word_2 = word_2.title()

#Se definen los condicionales que impriman el resultado según sea el caso.
if word_1 < word_2:
    print('\n'f'{word_1} está primero en el diccionario que {word_2}')
    
elif word_1 > word_2:
     print('\n'f'{word_2} está primero en el diccionario que {word_1}')
     
elif word_1 == word_2:
    print('\n'f'{word_2} y {word_1} son iguales.')
    
```

```python
"""14. Leer una cadena de texto y organice alfabéticamente cada una de las letras 
que la componen, repitiendo cada una tantas veces como se encuentra. Por 
ejemplo, la cadena ‘tarea importante’ será ‘aaaeeimnoprrttt’. (Note que no se 
incluyen los espacios)."""

#definición y salida en pantalla mensaje de bienvenida al programa.                                                               
welc = 'BIENVENIDO AL PROGRAMA ORGANIZADOR DE CARACTERES.'

print(welc.center(100))

#Lectura de datos desde el teclado. Se pide al usuario ingresar dos palabras.
#Se define una cadena de caracteres permitidos.
#Si estas palabras contienen caracteres que no están en la cadena de caracteres 
#permitidos se pide que sea ingresada de nuevo.
while True:
    try:                                                                 
        s = input('Ingrese una cadena de texto: ')
        car_perm = set('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUV\
                            WXYZ ')
        if set(s).issubset(car_perm):
            break
        else:
            print('Solo puede ingresar letras en la cadena')
    except ValueError:
        print ('Ingrese de nuevo la cadena')


lista_s = list(s)    #la cadena de texto s se transforma en lista (lista_s)

lista_s.sort()       #Mediante el comando sort se ordenan los elementos de la
#Lista_s

posicion=0           

#Se itera a través de lista_s
#Se eliminan los espacios de la lista.

while posicion < len(lista_s):
    
    if lista_s[posicion] == ' ':
        lista_s.pop(posicion)
    else:
        posicion=posicion+1
        break 

#li es la lista_s que vuelve a transformarse en cadena.
li=''.join(lista_s)

#Salida en pantalla de los caracteres ordenados.
print(f'Los caracteres de la cadena, organizados alfabéticamente, son: \n{li}')
```
