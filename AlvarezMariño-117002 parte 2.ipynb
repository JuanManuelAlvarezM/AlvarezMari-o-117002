```python
"""Leer una matriz cuadrada A, obtener su matriz triangular superior asociada U y 
determinar quÃ© fila de U contiene la mayor cantidad de unos."""

import numpy as np              #Se importa el módulo numpy

#Lectura de datos desde el teclado. Se pide al usuario ingresar la dimensión
#de una matriz cuadrada

while True:
    try:
        dim = int(input("Digite la dimensión de la matriz cuadrada: "))
        if dim > 1:
            break
        else: 
            print('la dimensión debe ser mayor a 1. Inténtelo de nuevo.')
    except ValueError:
        print('Debe ingresar un entero positivo. Inténntelo de nuevo.')
    
matrix = []      #se crea una lista vaciía que será la matriz al ingresar sus 
                 #elementos


#Lectura de datos desde el teclado. Se pide al usuario ingresar los elementos
#de la matriz.
for i in range(dim):          
    a = [] 
    for j in range(dim):       
         a.append(int(input('Digite los elementos de la matriz (fila por fila)\
: '))) 
         
    matrix.append(a) 

#Salida en pantalla de la matriz ingresada por el usuario
print('La matriz ingresada es')

for i in range(dim): 
    for j in range(dim): 
        print(matrix[i][j], end = " ") 
    print()

#Los elementos bajo la diagonal de la matriz se hacen 0.
for i in range(dim):
    for j in range (dim):
        if i > j:
            matrix[i][j]=0

#Se obtiene la matriz triangular superior U.
U=np.array(matrix)

#Salida en pantalla de la matriz U.
print('La matriz triangular superior asociada (U) es: ')

print(U)

max_1 =[0,0]   #Lista cuya primera entrada estarán la cantidad de unos y la 
#segunda en qué fila estarán

for i in range (dim):
    
    cont=0              #contador de unos
    
    for j in range(dim):
        
        if U[i][j]==1:
            
            cont += 1
            
        if cont > max_1[0]:
            max_1=[cont, i+1]
            
#Salida en pantalla de la fila con más unos de la matriz U.
if max_1[0]>0:
    print('la fila con mas unos (1) de la matriz U es la ', max_1[1])
elif max_1[0]==0:
    print('No hay unos (1) en la matriz U.')
```
